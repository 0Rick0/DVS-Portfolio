# Visualise some aspects of the survey

```python
# Some imports
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
# load dataset
survey = pd.read_csv('DVS-Portfolio/group_survey/GroupSurvey_Responses.csv')

# display head
survey.head()
```

```python
import pandas as pd
divide = survey['What is your main study profile?'].value_counts()  # type: pd.Series
ax1 = plt.subplot(121, aspect='equal')
divide.plot(kind='pie', ax=ax1, autopct='%1.1f%%', startangle=45, shadow=True, legend=False,
            fontsize=14, figsize=(20, 20))
plt.ylabel('')
plt.title('Divide between different profiles', fontsize=28)

interest = survey['What aspect of Data Visualization is most interesting to you?'].value_counts()  # type: pd.Series
ax2 = plt.subplot(122, aspect='equal')
interest.plot(kind='pie', ax=ax2, autopct='%1.1f%%', startangle=45, shadow=True, legend=False,
              fontsize=14, figsize=(20, 20))
plt.ylabel('')
plt.title('Interest', fontsize=28)

plt.show()
```
